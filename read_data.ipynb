```python
import pandas as pd
from datetime import datetime
df = pd.read_json('./data.json')
```

```python
def conv_to_datetime(series):
    for lst in series:
        for t_e in lst:
            t_e[0] = pd.to_datetime(t_e[0], unit='ms')
```

```python
conv_to_datetime(df['esg'])
conv_to_datetime(df['esg_industry'])
df
```
